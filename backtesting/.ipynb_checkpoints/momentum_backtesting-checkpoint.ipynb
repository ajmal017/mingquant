```python
%matplotlib inline
import matplotlib.pylab as plt
plt.rcParams['figure.dpi'] = 200

import pandas as pd
import os
import requests 
import pandas as pd
import requests
from yahooquery import Ticker

# import backtesting and yfinance

from backtesting import Backtest, Strategy
from backtesting.lib import crossover
from backtesting.test import SMA
import yfinance as yf
```

```python
# define backtesting strategy...simple SMA crossover
# see https://kernc.github.io/backtesting.py/doc/examples/Quick%20Start%20User%20Guide.html

# more strategies avalable!!!
# https://kernc.github.io/backtesting.py/doc/examples/Strategies%20Library.html

class SmaCross(Strategy):
    n1 = 10
    n2 = 20

    def init(self):
        self.sma1 = self.I(SMA, self.data.Close, self.n1)
        self.sma2 = self.I(SMA, self.data.Close, self.n2)

    def next(self):
        if crossover(self.sma1, self.sma2):
            self.buy()
        elif crossover(self.sma2, self.sma1):
            self.sell()
            
            
# generate dataset compatible with backtesting strategy

def get_data(tickerSymbol, startdate='2010-1-1', enddate='2020-7-7'):
    # tickerSymbol = "NVTA"
    tickerData = yf.Ticker(tickerSymbol)
    tdf = tickerData.history(start=startdate, end=enddate).drop(['Dividends', 'Stock Splits'], axis=1)
    tdf.index.name = None
    return tdf

# run the backtest and plot the results

def backtest_results(ticker, start, end, quiet=False):
    tdf = get_data(ticker,startdate=start, enddate=end)
    bt = Backtest(tdf, SmaCross, cash=10000, commission=.002)
    output = bt.run()
    if not quiet:
        print(output)
        bt.plot()
    return output

def get_mean_revgrowth(ticker, startdate='2015-07-01', enddate='2020-07-07'):
    temp = Ticker(ticker)
    df = temp.income_statement()
    df['date'] = pd.to_datetime(df['asOfDate'])
    mask = (df['date'] > startdate) & (df['date'] <= enddate)
    df = df.copy().loc[mask]
    df['revenue_growth'] = df['TotalRevenue'].pct_change()
    mean_rev_growth = df['revenue_growth'].mean()
    mean_revenue = df['TotalRevenue'].mean()
    n_statements = len(df)
    return mean_revenue, mean_rev_growth, n_statements

def days_since_cross(tickerSymbol, startdate='2010-1-1', enddate='2020-7-7'):
    tickerData = yf.Ticker(tickerSymbol)
    df = tickerData.history(period="1d", start=startdate, end=enddate)
    df['SMA_50'] = df.iloc[:,1].rolling(window=50).mean()
    df['SMA_250'] = df.iloc[:,1].rolling(window=250).mean()
    xdf = df[df['SMA_250'] < df['SMA_50']]
    ydf = df[df['SMA_250'] > df['SMA_50']]
    try:
        under = pd.to_datetime(ydf.iloc[:,-1].index)[-1]
        cross = pd.to_datetime(xdf.iloc[:,-1].index)[-1]
        xdays = int(str(cross-under).split()[0])
    except:
        xdays=0
    return xdays
```

```python
symbola = pd.read_csv("data/large_cap_volume_leaders.csv").Symbol
symbolb = pd.read_csv("data/nasdaq_volume_leaders.csv").Symbol
symbolc = pd.read_csv("data/mid_cap_volume_leaders.csv").Symbol
symbold = pd.read_csv("data/small_cap_volume.csv").Symbol
symbols = [i for i in set(symbola.append(symbolb).append(symbolc).append(symbold))]
print(len(symbols))
```

```python
startdate='2019-07-01'
enddate='2020-07-07'
```

```python
def get_company_stats(symbols, startdate, enddate):
    masterdf = pd.DataFrame()
    # stocks = [i for i in subdf.stock]
    for ticker in symbols:
        try:
            rdf = pd.DataFrame({'stock': [ticker]})
            rdf['stdev'] = yf.Ticker(ticker).history(start=startdate, end=enddate).drop(['Dividends', 'Stock Splits'], axis=1).Close.std()
            rdf['mean'] = yf.Ticker(ticker).history(start=startdate, end=enddate).drop(['Dividends', 'Stock Splits'], axis=1).Close.mean()
            rdf['stdev_mean'] = rdf['stdev']/rdf['mean']
            mean_revenue, mean_rev_growth, n_statements = get_mean_revgrowth(ticker, startdate, enddate)
            rdf['mean_revenue'] = mean_revenue
            rdf['mean_rev_growth'] = mean_rev_growth
            rdf['n_statements'] = n_statements
            rdf['days_since_cross'] = days_since_cross(ticker, startdate, enddate)
            masterdf = pd.concat([masterdf.copy(), rdf])
        except:
            print("failed to get company stats for: " + ticker)
    return masterdf
stock_stats = get_company_stats(symbols, startdate, enddate)
stock_stats.head()
```

```python
def get_backtest_results(symbols, startdate, enddate):
    masterdf = pd.DataFrame()
    # stocks = [i for i in subdf.stock]
    for ticker in symbols:
        try:
            results = backtest_results(ticker, startdate, enddate, quiet=True)
            rdf = pd.DataFrame(results).T
            rdf['advantage'] = rdf['Return [%]']/rdf['Buy & Hold Return [%]']
            rdf['symbol'] = ticker
            masterdf = pd.concat([masterdf.copy(), rdf])
        except:
            print("failed to get backtest results for: "+ ticker)
    return masterdf

masterdf = get_backtest_results(symbols, startdate, enddate)
# masterdf.to_csv("data/backtest_all_results.txt", sep="\t", index=False)
masterdf.head()
```
